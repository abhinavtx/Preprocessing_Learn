```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df.nunique()
```

```python
### One Hot Encoding using Normal Pandas
dfOneHot =  pd.get_dummies(df,columns=['fuel','owner','brand'])
dfOneHot.iloc[:,2:] = dfOneHot.iloc[:,2:].astype(int)
dfOneHot.head()
```

```python
### One Hot Encoding using Normal Pandas with dropping val
dfOneHot =  pd.get_dummies(df,columns=['fuel','owner','brand'], drop_first = True)
dfOneHot.iloc[:,2:] = dfOneHot.iloc[:,2:].astype(int)
dfOneHot.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',dtype=np.int32)
```

```python
X_new = ohe.fit_transform(df[['fuel','owner']]).toarray()
newMat = pd.DataFrame(X_new,columns=ohe.get_feature_names_out(['fuel', 'owner']))
newMat
```

```python
## Now using One Hot Encoder using pandas with top categories
thresh = df.shape[0]/100
print(thresh)
counts = df['brand'].value_counts()
```

```python
repl = counts[counts <= thresh].index
## repl is boolean mask which means it gives false if less than thresh and gives it in others car brand
```

```python
carBrand = pd.get_dummies(df['brand'].replace(repl, 'Other Brands'))
```

```python
carBrand = carBrand.astype(int)
carBrand
```
