```python
# Applying lambda functions
df['Salary'] = df['Salary'].apply(lambda num : num / 1000)
df.head()
```

```python
# Operation on every row, doing on a dictionary
# Function to add
def add_values(row):
    return row['A'] + row['B'] + row['C']
 
# Create a dictionary with three fields each
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Apply the user-defined function to every row
df['add'] = df.apply(add_values, axis=1)
df['average'] = df['add'].apply(lambda numb : numb / df.shape[1])
print('\nAfter Applying Function: ')
# Print the new DataFrame
print(df)
```

```python
def normalize(x, y):
    x_new = ((x - np.mean([x, y])) / (max(x, y) - min(x, y)))
    return x_new

df['normal'] = df.apply(lambda row: normalize(row['add'], row['average']), axis=1)
print('After Applying normalisation: \n', df)
```
