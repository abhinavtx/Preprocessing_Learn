```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('nba.csv', index_col='Name') # Mentoned the index column to be name
df.head(10) # Return initial top 10 rows
```

```python
df.tail(10) # Last 10 rows
# Note -> head() and tail() method are also valid for pd.series()
```

```python
#Pandas describe() is used to view some basic statistical details 
#like percentile, mean, std, etc. of a data frame or a series of numeric values
# Now for describe, we can do it individually for columns, like:
print(df['Age'].describe(), "\n" )
df.describe()
```

```python
# Making the copy of pandas dataframe
duplDf = df.copy(deep=True) # Use parameter of true or false to mention copy type
duplDf.drop(['Weight'], axis = 1, inplace= True) 
duplDf
# In this we have dropped a column from the deep copy of a dataframe
```

```python
# Accessing an item using boolean index
# Means accessing of rows based on true or false
df = pd.read_csv('nba.csv')
data = pd.DataFrame(df, index = [1,2,3,4,5,6])
mask = data.index > 4
print(data[mask])
```
