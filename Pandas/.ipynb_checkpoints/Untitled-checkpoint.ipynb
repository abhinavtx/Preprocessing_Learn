```python
import numpy as np
import pandas as pd
```

```python
dict1 = { 'SID' : [i for i in range(1,len(dict1['Name'])+1)] ,'Name':['Martha', 'Tim', 'Rob', 'Georgia'], 
        'Maths':[87, 91, 97, 95], 
        'Science':[83, 99, 84, 76] 
       } 
dict2 = {'SID' : [i+ len(dict1['Name']) for i in range(1,len(dict2['Name']) +1)] , 'Name':['Amy', 'Maddy'], 
        'Maths':[89, 90], 
        'Science':[93, 81] 
       } 
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
# By the use of ignore index = True 
df1
```

```python
newDf = pd.concat([df1,df2], ignore_index= True)
newDf = newDf.assign(Average = lambda x : (x['Maths'] + x['Science'])/2)
newDf
```

```python
newDf.sort_values(['Average', 'SID'], axis=0, ascending=[False, True], inplace=True)
# In this the order of writing them will be used for sorting them
newDf
```
