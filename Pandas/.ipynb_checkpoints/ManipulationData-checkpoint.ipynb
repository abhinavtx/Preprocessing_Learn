#### Manipulation of data 

```python
# Adding a column
import pandas as pd
import numpy as np
# Define a dictionary containing Students data
data = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],'Height': [5.1, 6.2, 5.1, 5.2],'Qualification': ['Msc', 'MA', 'Msc', 'Msc']}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
df.set_index("Name", inplace= True)
# Using DataFrame.insert() to add a column --> Age
df.insert(2, "Age", [21, 23, 24, 21])

# Observe the result
print(df)
```

```python
# Now inserting multiple columns
state = ['NCT', 'Karnataka', 'Tamil Nadu', 'Bihar']
city = ['Delhi', 'Bangalore', 'Chennai', 'Patna']
newData = {'City' : city ,'State' : state }
df.assign(**newData)
# NOTE -> Here ** is used to add multiple columns
```

```python
datag = pd.read_csv('nba.csv', index_col='Name')
datag
```

```python
datag.sort_values("Name", axis= 0, ascending= True, inplace= True, na_position='last')
datag.dropna(how = 'all')
```

```python
## Dropping rows where name is null
datag.dropna(subset=['Name'], inplace= True)
```

```python
datag.isna().sum()
```

```python
sal = datag['Salary'].mean()
datag['Salary'].fillna(value=sal, inplace=True)
```

```python
datag.isna().sum()
```

```python
datag
```

```python
df = datag.loc[:,["Number", "Age", "Weight", "Salary"]]
x = datag.loc[:,["Number", "Age", "Weight", "Salary"]]
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Choose a correlation threshold (absolute value)
correlation_threshold = 0.7

# Find pairs of columns with correlation below the threshold
correlated_pairs = [(col1, col2) for col1 in correlation_matrix.columns for col2 in correlation_matrix.columns
                    if abs(correlation_matrix.loc[col1, col2]) >= correlation_threshold and col1 != col2]

# Remove one column from each correlated pair
for col1, col2 in correlated_pairs:
    # Decide which column to remove based on your criteria
    # For simplicity, removing the second column in each pair
    df.drop(col2, axis=1, inplace=True)

# Display the DataFrame after removing correlated columns
print("\nDataFrame after removing correlated columns:")
print(x.info)
```
