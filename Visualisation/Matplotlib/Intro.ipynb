```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'g--') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
## For  plotting multiple plots on same cell ---> use plt.subplot()
plt.subplot(1,3,1)
## Here it means the arguements are number of rows, number of columns and index (here it follows 1 based indexing) respectively
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(1,3,3)
plt.plot(y, x, 'g--')
```
